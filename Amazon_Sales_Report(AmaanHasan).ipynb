***Insight 2 : Most Popular Product is T-Shirt***

```python
pivot_table = df.pivot_table(index='Category', columns='ship-country', values='Qty', aggfunc='sum', fill_value=0)

plt.figure(figsize=(12, 8))  # Adjust figure size as needed
sns.heatmap(pivot_table, annot=True, fmt="d", cmap="YlGnBu", linewidths=0.5)
plt.title('Qty')
plt.xlabel('ship-country')
plt.ylabel('Category')
plt.show()
```

***Insight 3 : Month with Highest Average sales***

```python
# Ensure 'Date' column is in datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y')

# Create a new column 'month' with the month names
df['month'] = df['Date'].dt.month_name()

# Calculate the average amount sold per month
avg_amount_per_month = df.groupby('month')['Amount'].mean()

# Sort by calendar month order if needed
month_order = ["January", "February", "March", "April", "May", "June", 
               "July", "August", "September", "October", "November", "December"]
avg_amount_per_month = avg_amount_per_month.reindex(month_order)

# Plot the line chart
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(avg_amount_per_month.index, avg_amount_per_month.values, marker='o', linestyle='-', color='b')
plt.title('Average Amount Sold per Month')
plt.xlabel('Month')
plt.ylabel('Average Amount Sold')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

***Insight 4 : Analysis of Fulfilment type along with status of delivery***

```python
import matplotlib.pyplot as plt

# Count occurrences of each Fulfilment and Status combination
status_fulfilment_counts = df.groupby(['Fulfilment', 'Status']).size().unstack()

# Plot a stacked bar chart
status_fulfilment_counts.plot(kind='bar', stacked=True, figsize=(10, 6), colormap="Set1")
plt.title('Distribution of Status by Fulfilment Type')
plt.xlabel('Fulfilment')
plt.ylabel('Count')
plt.legend(title='Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

***Insight 5 : State with the most number of orders***

```python
# Calculate the total quantity shipped per state
quantity_per_state = df.groupby('ship-state')['Qty'].sum().sort_values(ascending=False)

# Plot the bar plot
plt.figure(figsize=(20, 6))
sns.barplot(x=quantity_per_state.index, y=quantity_per_state.values, palette="Set2")
plt.title('Total Quantity Shipped per State')
plt.xlabel('ship-state')
plt.ylabel('Qty')
plt.xticks(rotation=90, ha='right')
plt.show()
```
